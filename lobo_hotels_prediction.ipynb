```python
predictions = pd.read_csv('/content/SLM_S2021/Competition/hotels_test.csv')
```

```python
predictions = predictions.drop(['previous_bookings_not_canceled', 'deposit_type', 'meal'], axis =1)
```

```python
predictions = predictions.drop(['meal'], axis =1)
```

```python
predictions['stays_in_nights'] = predictions['stays_in_weekend_nights'] +  predictions['stays_in_week_nights']
```

```python
predictions = predictions.replace({"Resort Hotel": 0, "City Hotel":1})
predictions['market_segment'] = predictions['market_segment'].replace({'Direct': 0, 'Corporate': 1, 'Online TA':2, 'Offline TA/TO':3,
       'Complementary':4, 'Groups':5, 'Aviation':6})
predictions['arrival_date_month'] = predictions['arrival_date_month'].replace({'January': 1, 'February': 2,'March': 3,'April': 4,'May': 5,'June': 6,'July': 7,'August': 8,
            'September': 9, 
            'October': 10,
            'November': 11, 'December': 12 })
predictions['distribution_channel'] = predictions['distribution_channel'].replace({'Direct':0, 'Corporate':1, 'TA/TO':2, 'Undefined':3, 'GDS':4})
predictions['meal'] = predictions['meal'].replace({'BB': 1, 'HB': 2,'SC': 3,'Undefined': 4,'FB': 5})
predictions['deposit_type'] = predictions['deposit_type'].replace({'No Deposit':0, 'Refundable':1, 'Non Refund':2})
predictions['customer_type'] = predictions['customer_type'].replace({'Transient':0, 'Contract':1, 'Transient-Party':2, 'Group':3})
predictions['reserved_room_type'] = predictions['reserved_room_type'].replace({'A': 1, 'B': 2,'C': 3,'D': 4,'E': 5,'F': 6,'G': 7,'H': 8,
            'L': 9, 
            'P': 10 })
predictions['assigned_room_type'] = predictions['assigned_room_type'].replace({ 'A' : 1 , 'D' : 2, 'E' : 3, 'C' : 4, 'F':5, 'G':6, 'B':7, 'H':8, 'I': 9, 'K':10, 'P':11, 'L':12})
```

```python
results = clf.predict(std_scaler.transform(predictions))
```

```python
res= pd.DataFrame(results, columns = ['Prediction'])
res.to_csv('lobo_Hotels_prediction.csv') 
```
