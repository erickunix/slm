```python
import pandas as pd
import sklearn
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/SLM_S2021/Competition/hotels_train.csv')
df
```

```python
df.describe()
```

```python
df.head()
```

```python
df = df.replace({"Resort Hotel": 0, "City Hotel":1})
```

```python
df['market_segment'].unique()
```

```python
df['market_segment'] = df['market_segment'].replace({'Direct': 0, 'Corporate': 1, 'Online TA':2, 'Offline TA/TO':3,
       'Complementary':4, 'Groups':5, 'Aviation':6})
```

```python
df['arrival_date_month'] = df['arrival_date_month'].replace({'January': 1, 'February': 2,'March': 3,'April': 4,'May': 5,'June': 6,'July': 7,'August': 8,
            'September': 9, 
            'October': 10,
            'November': 11, 'December': 12 })
```

```python
df['distribution_channel'].unique()
```

```python
df['distribution_channel'] = df['distribution_channel'].replace({'Direct':0, 'Corporate':1, 'TA/TO':2, 'Undefined':3, 'GDS':4})
```

```python
df['meal'] = df['meal'].replace({'BB': 1, 'HB': 2,'SC': 3,'Undefined': 4,'FB': 5})
```

```python
df['deposit_type'].unique()
```

```python
df['deposit_type'] = df['deposit_type'].replace({'No Deposit':0, 'Refundable':1, 'Non Refund':2})
```

```python
df['customer_type'].unique()
```

```python
df['customer_type'] = df['customer_type'].replace({'Transient':0, 'Contract':1, 'Transient-Party':2, 'Group':3})
```

```python
df['reserved_room_type'] = df['reserved_room_type'].replace({'A': 1, 'B': 2,'C': 3,'D': 4,'E': 5,'F': 6,'G': 7,'H': 8,
            'L': 9, 
            'P': 10 })
```

```python
df['assigned_room_type'] = df['assigned_room_type'].replace({ 'A' : 1 , 'D' : 2, 'E' : 3, 'C' : 4, 'F':5, 'G':6, 'B':7, 'H':8, 'I': 9, 'K':10, 'P':11, 'L':12})
```

```python
y = df['is_canceled']
x = df.drop(['is_canceled'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0).fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
clf.score(x_train, y_train)
```
